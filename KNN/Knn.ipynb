```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('D:\Machine learning\KNN\Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x= df.iloc[:, :-1]
y= df.iloc[:, -1]
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', palette='viridis', data=df)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
%pip install scikit-learn
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=16)
knn.fit(x_train, y_train)
```

```python
print("Accuracy on training set:", knn.score(x_train, y_train)*100)
```

```python
print("Accuracy on test set:", knn.score(x_test, y_test)*100)
```

```python
for i in range(1, 30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    print(f"Accuracy for k={i}: {knn.score(x_test, y_test)*100:.2f}%")
    print(f"Accuracy for k={i}: {knn.score(x_train, y_train)*100:.2f}%")
```

```python
knn.predict([[1.083596,-0.990844]])
```

```python
plot_decision_regions(x.values, y.values, clf=knn, legend=2)
plt.show()
```
