```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv(r'D:\Machine learning\Naive_Bayes\college_placement_dataset_200.csv')
```

```python
df.head(3)
```

```python
sns.kdeplot(df['score'], shade=True)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='cgpa', y='score', hue='placed', data=df)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = df.iloc[:,:-1]
y = df['placed']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
gnb = GaussianNB()
```

```python
gnb.fit(x_train, y_train)
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=gnb, legend=2)
plt.show()
```

```python
gnb.score(x_test, y_test)*100
```

```python
gnb.predict([[6.87,79.0	]])
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=mnb, legend=2)
plt.show()
```

```python
mnb.score(x_test, y_test)*100
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=bnb, legend=2)
plt.show()
```

```python
bnb.score(x_test, y_test)*100
```
